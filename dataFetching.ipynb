```python
id_list = []
for track in tks:
    idtemp = track['track']['id']
    id_list.append(idtemp)
len(id_list)    
```

```python
allFeatures = []
while len(id_list):
    
    if len(id_list) >= 100:
        tempList = id_list[0:100]
        #print(len(id_list))
        #print(len(tempList))
        f = sp.audio_features(tempList)
        allFeatures.extend(f)
        id_list = id_list[100:]
    else:
        f = sp.audio_features(id_list)
        allFeatures.extend(f)
        id_list = []
        
len(allFeatures)
```
